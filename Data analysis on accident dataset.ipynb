## Data Visualization and Statistical Analysis

```python
df.info()
```

```python
df.describe()
```

```python
print(df['accident_year'].value_counts())

# Casualty Class
print(df['casualty_class'].value_counts())
```

```python
print(pd.crosstab(df['casualty_class'], df['casualty_severity']))
```

```python
# Casualty Severity Analysis
severity_counts = df['casualty_severity'].value_counts()
print(severity_counts)
```

```python
plt.figure()
df['casualty_severity'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Casualty Severity Distribution')
plt.ylabel('')
plt.show()
```

```python
# Age Distribution
age_distribution = df['age_of_casualty'].describe()
print(age_distribution)
```

```python
plt.figure()
sns.histplot(df['age_of_casualty'], bins=20, kde=True, color='purple')
plt.title('Age Distribution of Casualties', fontsize=16)
plt.xlabel('Age of Casualty', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure()
sns.boxplot(x='casualty_severity', y='age_of_casualty', data=df, palette='coolwarm')
plt.title('Age of Casualty by Severity', fontsize=16)
plt.xlabel('Casualty Severity', fontsize=14)
plt.ylabel('Age of Casualty', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Casualty Class
class_counts = df['casualty_class'].value_counts()
print(class_counts)
```

```python
plt.figure()
sns.countplot(x='casualty_class', data=df)
plt.title('Distribution of Casualty Class')
plt.xlabel('Casualty Class')
plt.ylabel('Count')
plt.show()
```

```python
# Sex of Casualty
sex_counts = df['sex_of_casualty'].value_counts()
print(sex_counts)
```

```python
plt.figure(figsize=(10, 5))
df['sex_of_casualty'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Sex of Casualty Distribution')
plt.ylabel('')
plt.show()
```

```python
type_counts = df['casualty_type'].value_counts()
print(type_counts)
```

```python
# Countplot of casualty type
plt.figure()
sns.countplot(x='casualty_type', data=df)
plt.title('Distribution of Casualty Type')
plt.xlabel('Casualty Type')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot of casualty home area type
plt.figure()
sns.countplot(x='casualty_home_area_type', data=df)
plt.title('Distribution of Casualty Home Area Type')
plt.xlabel('Casualty Home Area Type')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot of car passenger
plt.figure()
sns.countplot(x='car_passenger', data=df)
plt.title('Distribution of Car Passenger')
plt.xlabel('Car Passenger')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.countplot(x='bus_or_coach_passenger', data=df)
plt.title('Distribution of Bus or Coach Passenger')
plt.xlabel('Bus or Coach Passenger')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot of pedestrian road maintenance worker
plt.figure()
sns.countplot(x='pedestrian_road_maintenance_worker', data=df)
plt.title('Distribution of Pedestrian Road Maintenance Worker')
plt.xlabel('Pedestrian Road Maintenance Worker')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.barplot(x='casualty_severity', y='age_of_casualty', data=df, ci=None, palette='viridis')
plt.title('Mean Age of Casualty by Severity', fontsize=16)
plt.xlabel('Casualty Severity', fontsize=14)
plt.ylabel('Mean Age of Casualty', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Accident by IMD Decile
imd_decile_counts = df['casualty_imd_decile'].value_counts()
print(imd_decile_counts)
```

```python
# Barplot of mean IMD decile by casualty severity
plt.figure()
sns.barplot(x='casualty_severity', y='casualty_imd_decile', data=df, ci=None)
plt.title('Mean IMD Decile by Casualty Severity')
plt.xlabel('Casualty Severity')
plt.ylabel('Mean IMD Decile')
plt.show()
```

```python
facet = sns.FacetGrid(df, col='casualty_severity', height=5, aspect=1.2)
facet.map(sns.histplot, 'age_of_casualty', bins=15, kde=True)
facet.set_titles(col_template='{col_name}')
facet.set_axis_labels('Age of Casualty', 'Count')
facet.fig.suptitle('Age Distribution by Casualty Severity', fontsize=16)
facet.fig.tight_layout()
facet.fig.subplots_adjust(top=0.85)
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='age_of_casualty', y='casualty_imd_decile', hue='casualty_severity', data=df, palette='Set1')
sns.regplot(x='age_of_casualty', y='casualty_imd_decile', data=df, scatter=False, color='blue')
plt.title('Scatter Plot with Regression Line: Age vs IMD Decile', fontsize=16)
plt.xlabel('Age of Casualty', fontsize=14)
plt.ylabel('IMD Decile', fontsize=14)
plt.legend(title='Casualty Severity')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure()
sns.pairplot(df[['age_of_casualty', 'casualty_imd_decile', 'casualty_severity']])
plt.title('Pairplot of Numerical Features')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
