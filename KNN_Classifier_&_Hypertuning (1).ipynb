```python
import pandas as pd
df= pd.read_csv("/content/bank-full.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df= df.drop(["contact","poutcome"], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['job'] = label_encoder.fit_transform(df['job'])
df['marital'] = label_encoder.fit_transform(df['marital'])
df['default'] = label_encoder.fit_transform(df['default'])
df['housing'] = label_encoder.fit_transform(df['housing'])
df['loan'] = label_encoder.fit_transform(df['loan'])
df['month'] = label_encoder.fit_transform(df['month'])
df['y'] = label_encoder.fit_transform(df['y'])
df['education'] = label_encoder.fit_transform(df['education'])


```

```python
df.head()
```

```python
x= df[['default','age','job','marital','education','balance','loan','day','month','duration','campaign','pdays','previous']]
y= df["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
per= model.predict(x_test)
per
```

```python
pd.DataFrame({"Actual": y_test, "Predictions": per})
```

## GridSearchCV

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': [3,5,7,9],
    'p': [1,2]  # 1 for Manhattan distance, 2 for Euclidean distance
}
```

```python
grid_search = GridSearchCV(model, param_grid)
```

```python
grid_search.fit(x_train, y_train)
print("Best Parameters: ", grid_search.best_params_)
```

```python
y_per = grid_search.predict(x_test)
y_per
```

```python
pd.DataFrame({"Actual": y_test, "Predictions": y_per})
```

## RandomSearchCV

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier()
```

```python
from sklearn.model_selection import  RandomizedSearchCV
param_grid= {'n_neighbors':[1,3,5,7],'metric': [1,2]}
```

```python
random_search= RandomizedSearchCV(model,param_grid, n_iter=3)
```

```python
random_search.fit(x_train, y_train)
random_search.best_params_
```

```python
y_per = random_search.predict(x_test)
y_per
```

```python
pd.DataFrame({"Actual": y_test, "Predictions": y_per})
```
