## Objective 1: Impact of Speed Limits on Frequency and Severity of Accidents

```python
speed_df = df[(df['speed_limit'] > 0) & (df['accident_severity'].notnull())]

# Speed limit vs Accident Severity
plt.figure()
sns.boxplot(x='speed_limit', y='accident_severity', data=speed_df)
plt.title('Impact of Speed Limits on Accident Severity')
plt.xlabel('Speed Limit')
plt.ylabel('Accident Severity')
plt.show()
```
