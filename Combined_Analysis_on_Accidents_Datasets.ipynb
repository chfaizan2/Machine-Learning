## Objective 1: Impact of Speed Limits on Frequency and Severity of Accidents

```python
speed_df = df[(df['speed_limit'] > 0) & (df['accident_severity'].notnull())]

# Speed limit vs Accident Severity
plt.figure()
sns.boxplot(x='speed_limit', y='accident_severity', data=speed_df)
plt.title('Impact of Speed Limits on Accident Severity')
plt.xlabel('Speed Limit')
plt.ylabel('Accident Severity')
plt.show()
```

```python
# Speed Limit vs Number of Casualties by Road Type
plt.figure(figsize=(12, 6))
sns.scatterplot(x='speed_limit', y='number_of_casualties', hue='road_type', data=speed_df)
plt.title('Speed Limit vs Number of Casualties by Road Type')
plt.xlabel('Speed Limit')
plt.ylabel('Number of Casualties')
plt.legend(title='Road Type')
plt.show()
```

```python
# Group by speed limit and count number of accidents
speed_limit_analysis = df.groupby('speed_limit').size().reset_index(name='accident_count')

# Plotting accident count by speed limit
plt.figure(figsize=(8,6))
barplot = sns.barplot(x='speed_limit', y='accident_count', data=speed_limit_analysis, palette='viridis')

# Adding data labels
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9), textcoords = 'offset points')

# Enhancing the plot
plt.title('Accident Count by Speed Limit', fontsize=16, weight='bold')
plt.xlabel('Speed Limit (mph)', fontsize=14)
plt.ylabel('Accident Count', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
