## Objective 1: Impact of Speed Limits on Frequency and Severity of Accidents

```python
speed_df = df[(df['speed_limit'] > 0) & (df['accident_severity'].notnull())]

# Speed limit vs Accident Severity
plt.figure()
sns.boxplot(x='speed_limit', y='accident_severity', data=speed_df)
plt.title('Impact of Speed Limits on Accident Severity')
plt.xlabel('Speed Limit')
plt.ylabel('Accident Severity')
plt.show()
```

```python
# Speed Limit vs Number of Casualties by Road Type
plt.figure(figsize=(12, 6))
sns.scatterplot(x='speed_limit', y='number_of_casualties', hue='road_type', data=speed_df)
plt.title('Speed Limit vs Number of Casualties by Road Type')
plt.xlabel('Speed Limit')
plt.ylabel('Number of Casualties')
plt.legend(title='Road Type')
plt.show()
```

```python
# Group by speed limit and count number of accidents
speed_limit_analysis = df.groupby('speed_limit').size().reset_index(name='accident_count')

# Plotting accident count by speed limit
plt.figure(figsize=(8,6))
barplot = sns.barplot(x='speed_limit', y='accident_count', data=speed_limit_analysis, palette='viridis')

# Adding data labels
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9), textcoords = 'offset points')

# Enhancing the plot
plt.title('Accident Count by Speed Limit', fontsize=16, weight='bold')
plt.xlabel('Speed Limit (mph)', fontsize=14)
plt.ylabel('Accident Count', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Objective 2: Impact of Weather Conditions on Accident Rates and Types

```python
# Bar Plot: Weather Conditions vs Accident Frequency by Vehicle Type
plt.figure()
sns.countplot(x='weather_conditions', hue='vehicle_type', data=df)
plt.title('Impact of Weather Conditions on Accident Frequency by Vehicle Type')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.legend(title='Vehicle Type')
plt.show()
```

```python
# Stacked Bar Plot: Weather Conditions vs Vehicle Type
weather_vehicle_stack = df.groupby(['weather_conditions', 'vehicle_type']).size().unstack().fillna(0)
weather_vehicle_stack.plot(kind='bar', stacked=True, figsize=(8,6), cmap='viridis')
plt.title('Accident Frequency by Weather Conditions and Vehicle Type')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.legend(title='Vehicle Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Group by weather condition and accident type
weather_type_analysis = df.groupby('weather_conditions')['casualty_type'].value_counts().unstack().fillna(0)

# Plotting accident type by weather condition
plt.figure(figsize=(12, 8))
weather_type_analysis.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Accident Type by Weather Condition', fontsize=16, weight='bold')
plt.xlabel('Weather Condition', fontsize=14)
plt.ylabel('Accident Count', fontsize=14)

# Adjust the legend position
plt.legend(title='Casualty Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

## Objective 3: Temporal Trends in Accident Occurrences

```python
# Calculate weekly accident counts
weekly_accidents = df.groupby(['day_of_week', 'urban_or_rural_area']).size().reset_index(name='number_of_accidents')
```

```python
# Line Plot: Daily Frequency of Accidents
plt.figure(figsize=(12, 6))
sns.lineplot(x='day_of_week', y='number_of_accidents', hue='urban_or_rural_area', data=weekly_accidents)
plt.title('Daily Frequency of Accidents by Urban/Rural Area')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.legend(title='Area Type')
plt.show()
```

```python
# Ensure the date column is in datetime format with dayfirst=True
df['date'] = pd.to_datetime(df['date'], dayfirst=True)

# Add columns for day of the week and month if not already present
df['day_of_week'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
```

```python
monthly_accidents = df.groupby(['month', 'urban_or_rural_area']).size().reset_index(name='number_of_accidents')

# Line Plot: Monthly Frequency of Accidents
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='number_of_accidents', hue='urban_or_rural_area', data=monthly_accidents)
plt.title('Monthly Frequency of Accidents by Urban/Rural Area')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.legend(title='Area Type')
plt.show()
```

## 4.Investigate the spatial patterns of accidents across different regions of the UK.

```python
# Bar Plot: Accidents by Region and Driver Age Band
plt.figure(figsize=(12, 6))
sns.countplot(x='local_authority_district', hue='age_band_of_driver', data=df)
plt.title('Accidents by Region and Driver Age Band')
plt.xlabel('Region')
plt.ylabel('Number of Accidents')
plt.legend(title='Driver Age Band')
plt.xticks(rotation=90)
plt.show()
```

```python
# Map Plot: Accident Locations
fig = px.scatter_mapbox(df, lat='latitude', lon='longitude', color='age_band_of_driver',
                        mapbox_style="carto-positron", zoom=5, title="Accident Locations by Driver Age Band")
fig.show()
```
