## Data Visualization and staistical Analysis

```python
sns.set(style="whitegrid")
plt.figure()
sns.boxplot(df['location_easting_osgr'], palette="Set2")
plt.title('Boxplot of Location Easting OSGR', fontsize=16)
plt.xlabel('Location Easting OSGR', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure()
sns.boxplot(df['location_easting_osgr'])
plt.show()
```

```python
plt.figure()
sns.boxplot(df['location_northing_osgr'])
plt.show()
```

```python
sns.set(style="whitegrid", palette="muted")
plt.figure()
sns.boxplot(x='longitude', data=df, color='skyblue')
plt.xlabel('Longitude', fontsize=14)
plt.title('Boxplot of Longitude Values', fontsize=16)
sns.despine(trim=True)
plt.show()
```

```python
plt.figure()
sns.boxplot(df['latitude'])
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.countplot(x='police_force', data=df, palette="Set2")
plt.title('Count of Incidents by Police Force', fontsize=16)
plt.xlabel('Police Force', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['accident_severity'], bins=20, kde=True)
plt.title('Distribution of accident severity')
plt.xlabel('accident severity')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure()
sns.countplot(x='weather_conditions', data=df, palette="viridis")
plt.title('Count of Incidents by Weather Conditions', fontsize=16)
plt.xlabel('Weather Conditions', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['light_conditions'], bins=20, kde=True)
plt.title('Distribution of lightconditions')
plt.xlabel('lightconditions')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.countplot(x='road_surface_conditions', data=df, palette='muted')
plt.xlabel('Road Surface Conditions', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Distribution of Accidents by Road Surface Conditions', fontsize=16)
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}',
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='baseline', fontsize=12, color='black',
                       xytext=(0, 5), textcoords='offset points')
plt.show()
```

```python
plt.figure()
sns.countplot(x='urban_or_rural_area', data=df, palette='Set2')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Count')
plt.title('Distribution of Accidents by Urban or Rural Area')
plt.show()
```

```python
plt.figure()
sns.histplot(df['did_police_officer_attend_scene_of_accident'], bins=20, kde=True)
plt.title('Distribution of did police officer attend scene of accident')
plt.xlabel('Distribution of did police officer attend scene of accident')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.countplot(x='trunk_road_flag', data=df, palette='Set3')
plt.xlabel('Trunk Road Flag')
plt.ylabel('Count')
plt.title('Distribution of Accidents by Trunk Road Flag')
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                       textcoords='offset points')
plt.show()
```
